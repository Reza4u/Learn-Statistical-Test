

### Overall Presentation Objective

To equip the audience with a clear framework for selecting the appropriate statistical test based on research questions, data types, and underlying assumptions, and to understand alternatives when assumptions are not met.

2.  **Subsequent Slides (Grouped by Research Question/Comparison Type):**
    *   Group them by the *type of comparison* they facilitate. and present a smooth way one by one.

    **A. Comparing Two Independent Groups**
    *   **Slide X: Comparing Two Independent Groups**
        *   **Research Question:** Is there a significant difference between two unrelated groups?
        *   **Decision Point:**
            *   **Parametric Option:** Student's t-test
                *   **Assumptions:** Interval/Ratio data, independence of observations, normality, homogeneity of variances.
                *   **When to Use:** Comparing means of two independent groups.
                *   **If Assumptions Violated (especially normality/homogeneity):**
                    *   **Alternative:** Mann-Whitney U test
                        *   **Assumptions:** Ordinal/Interval/Ratio data, independence of observations.
                        *   **When to Use:** Comparing medians (or distributions) of two independent groups when data are not normal or are ordinal.
    

    **B. Comparing Two Dependent (Paired) Groups**
    *   **Slide Y: Comparing Two Dependent Groups**
        *   **Research Question:** Is there a significant difference between two related measurements (e.g., pre/post, matched pairs)?
        *   **Decision Point:**
            *   **Parametric Option:** Paired t-test
                *   **Assumptions:** Interval/Ratio data, independence of pairs, normality of the *differences* between pairs.
                *   **When to Use:** Comparing means of two paired samples.
                *   **If Assumptions Violated (especially normality of differences):**
                    *   **Alternative:** Wilcoxon signed-rank test
                        *   **Assumptions:** Ordinal/Interval/Ratio data, independence of pairs.
                        *   **When to Use:** Comparing medians (or distributions) of two paired samples when data are not normal or are ordinal.

    **C. Comparing Three or More Independent Groups**
    *   **Slide Z: Comparing 3+ Independent Groups**
        *   **Research Question:** Is there a significant difference among three or more unrelated groups?
        *   **Decision Point:**
            *   **Parametric Option:** One-way ANOVA
                *   **Assumptions:** Interval/Ratio data, independence of observations, normality within groups, homogeneity of variances.
                *   **When to Use:** Comparing means of three or more independent groups.
                *   **If Assumptions Violated (especially normality/homogeneity):**
                    *   **Alternative:** Kruskal-Wallis H test
                        *   **Assumptions:** Ordinal/Interval/Ratio data, independence of observations.
                        *   **When to Use:** Comparing medians (or distributions) of three or more independent groups when data are not normal or are ordinal.

    **D. Comparing Three or More Dependent (Repeated Measures) Groups**
    *   **Slide A: Comparing 3+ Dependent Groups (Repeated Measures)**
        *   **Research Question:** Is there a significant difference across three or more related measurements (e.g., multiple time points, multiple conditions on the same subjects)?
        *   **Decision Point:**
            *   **Parametric Option:** Repeated Measures ANOVA
                *   **Assumptions:** Interval/Ratio data, independence of subjects, normality of differences, sphericity (less critical with modern software, but good to mention).
                *   **When to Use:** Comparing means across three or more related samples.
                *   **If Assumptions Violated (especially normality/sphericity):**
                    *   **Alternative:** Friedman test
                        *   **Assumptions:** Ordinal/Interval/Ratio data, independence of subjects.
                        *   **When to Use:** Comparing medians (or distributions) across three or more related samples when data are not normal or are ordinal.
            *   **More Complex/Flexible Option:** Mixed-effect Regression
                *   **When to Use:** When you have hierarchical or clustered data, repeated measures with missing data, or want to model individual trajectories and group effects simultaneously. It's very flexible and can handle various data types and distributions.
                *   **Assumptions:** Depends on the specific model (e.g., linearity, independence of residuals, appropriate distribution for the outcome).
                *   **Note:** This is a more advanced topic, so keep the explanation concise and focus on its utility for complex repeated measures or clustered data.

    **E. Analyzing Categorical Data (Frequencies/Proportions)**
    *   **Slide B: Analyzing Categorical Data**
        *   **Research Question:** Are observed frequencies/proportions different from expected, or is there an association between two categorical variables?
        *   **Decision Points:**
            *   **For Independent Categorical Data:**
                *   **Chi-square test:**
                    *   **Assumptions:** Independent observations, expected cell counts > 5 (or a reasonable proportion).
                    *   **When to Use:** Testing association between two categorical variables (e.g., gender and preference), or goodness-of-fit for one categorical variable.
                *   **Fisher's exact test:**
                    *   **Assumptions:** Independent observations.
                    *   **When to Use:** Preferred over Chi-square when expected cell counts are very low (especially in 2x2 tables).
            *   **For Dependent Categorical Data:**
                *   **McNemar's test:**
                    *   **Assumptions:** Paired nominal data.
                    *   **When to Use:** Comparing proportions in paired samples (e.g., before/after intervention on a binary outcome).

### Slide Design Tips

*   **Keep it Clean:** Minimal text, maximum impact.
*   **Visuals are Key:**
    *   **Flowcharts/Decision Trees:** Absolutely essential for this topic. Build them step-by-step on the slide or show the full one and then zoom into sections.
    *   **Tables:** For comparing parametric vs. non-parametric, or assumptions.
    *   **Simple Diagrams:** To illustrate concepts like independent vs. paired.
*   **Consistent Formatting:** Use the same fonts, colors, and layout throughout.
*   **Use Icons:** To represent concepts like "groups," "time," "data type."
*   **Examples (Optional but Recommended):** For each test, a very brief, relatable example can solidify understanding. E.g., "Student's t-test: Comparing average test scores of two different teaching methods."

### How to Make it Better for Understanding

1.  **Start with the "Why":** Emphasize the consequences of choosing the wrong test.
2.  **Build a Decision Tree/Flowchart:** This is the single most effective visual for this topic. Start simple and add complexity.
3.  **Focus on the "When" and "Why":** Instead of just listing tests, explain the scenarios where each is appropriate.
4.  **Highlight Assumptions and Alternatives:** This is your core strength. Make it clear that violating assumptions doesn't mean the end of the world, but requires a different approach.
5.  **Use Consistent Language:** For assumptions, data types, etc.
6.  **Interactive Elements (if possible):** If it's a smaller group, you could pose a simple scenario and ask the audience which test they'd choose.

By structuring your presentation around the decision-making process and using strong visuals like flowcharts, you'll create a highly effective and memorable guide for choosing statistical tests. Good luck!

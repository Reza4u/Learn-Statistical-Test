## Independent (un-paired) Tests

এই পরীক্ষাগুলো ব্যবহৃত হয় যখন বিভিন্ন গ্রুপের ডেটা পয়েন্টগুলো একে অপরের থেকে স্বাধীন থাকে।

### 1. Student's t-test
*   **উদ্দেশ্য (Purpose):** দুটি স্বাধীন গ্রুপের গড় (mean) তুলনা করা。
*   **Assumptions (ধারণাসমূহ):**
    1.  **স্বাধীন পর্যবেক্ষণ (Independent observations):** প্রতিটি গ্রুপের পর্যবেক্ষণ স্বাধীন এবং গ্রুপগুলোর মধ্যেও পর্যবেক্ষণ স্বাধীন হতে হবে。
    2.  **নরমালিটি (Normality):** প্রতিটি গ্রুপের ডেটা মোটামুটিভাবে নরমাল ডিস্ট্রিবিউশন অনুসরণ করবে। (Shapiro-Wilk test বা Q-Q plot দিয়ে পরীক্ষা করা যায়)。
    3.  **ভেরিয়েন্সের সমতা (Homogeneity of variances / Homoscedasticity):** দুটি গ্রুপের ভেরিয়েন্স প্রায় সমান হতে হবে। (Levene's test দিয়ে পরীক্ষা করা যায়)。
    4.  **পরিমাপের স্কেল (Scale of measurement):** ডিপেন্ডেন্ট ভেরিয়েবল কন্টিনিউয়াস (continuous) হতে হবে (interval বা ratio স্কেল)。
*   **যদি Assumptions লঙ্ঘিত হয়:**
    *   নরমালিটি লঙ্ঘিত হলে: **Mann-Whitney U test** (non-parametric বিকল্প)。
    *   ভেরিয়েন্সের সমতা লঙ্ঘিত হলে: **Welch's t-test** (এটি Student's t-test এর একটি সংস্করণ যা ভেরিয়েন্সের অসমতা অ্যাডজাস্ট করে)。
    *   স্বাধীন পর্যবেক্ষণ লঙ্ঘিত হলে: এটি ডিজাইন বা ডেটা সংগ্রহের সমস্যা নির্দেশ করে। এর জন্য ভিন্ন মডেলিং কৌশল (যেমন, mixed-effects models) প্রয়োজন হতে পারে।

### 2. One-way ANOVA (Analysis of Variance)
*   **উদ্দেশ্য (Purpose):** তিন বা ততোধিক স্বাধীন গ্রুপের গড় (mean) তুলনা করা。
*   **Assumptions (ধারণাসমূহ):**
    1.  **স্বাধীন পর্যবেক্ষণ (Independent observations):** প্রতিটি গ্রুপের পর্যবেক্ষণ স্বাধীন এবং গ্রুপগুলোর মধ্যেও পর্যবেক্ষণ স্বাধীন হতে হবে。
    2.  **নরমালিটি (Normality):** প্রতিটি গ্রুপের ডেটা (বা residuals) মোটামুটিভাবে নরমাল ডিস্ট্রিবিউশন অনুসরণ করবে。
    3.  **ভেরিয়েন্সের সমতা (Homogeneity of variances):** সকল গ্রুপের ভেরিয়েন্স প্রায় সমান হতে হবে। (Levene's test বা Bartlett's test দিয়ে পরীক্ষা করা যায়)。
    4.  **পরিমাপের স্কেল (Scale of measurement):** ডিপেন্ডেন্ট ভেরিয়েবল কন্টিনিউয়াস (continuous) হতে হবে。
*   **যদি Assumptions লঙ্ঘিত হয়:**
    *   নরমালিটি বা ভেরিয়েন্সের সমতা লঙ্ঘিত হলে: **Kruskal-Wallis H test** (non-parametric বিকল্প)。
    *   শুধুমাত্র ভেরিয়েন্সের সমতা লঙ্ঘিত হলে (কিন্তু নরমালিটি থাকলে): **Welch's ANOVA**।

### 3. Mann-Whitney U test (Wilcoxon rank-sum test)
*   **উদ্দেশ্য (Purpose):** দুটি স্বাধীন গ্রুপের ডেটার ডিস্ট্রিবিউশন (distribution) বা মধ্যমা (median) তুলনা করা। এটি Student's t-test এর non-parametric বিকল্প, বিশেষত যখন নরমালিটির ধারণা লঙ্ঘিত হয়。
*   **Assumptions (ধারণাসমূহ):**
    1.  **স্বাধীন পর্যবেক্ষণ (Independent observations):** দুটি গ্রুপ স্বাধীন হতে হবে。
    2.  **পরিমাপের স্কেল (Scale of measurement):** ডিপেন্ডেন্ট ভেরিয়েবল অন্তত অর্ডিনাল (ordinal) স্কেলে পরিমাপযোগ্য হতে হবে。
    3.  **ডিস্ট্রিবিউশনের আকার (Shape of distribution):** যদি মধ্যমার তুলনা করতে চান, তবে দুটি গ্রুপের ডিস্ট্রিবিউশনের আকার এবং বিস্তার (shape and spread) একই রকম হতে হবে। অন্যথায়, এটি সাধারণভাবে ডিস্ট্রিবিউশনের মধ্যে পার্থক্য পরীক্ষা করে。
*   **বিকল্প:** এটি নিজেই t-test এর নরমালিটি assumption লঙ্ঘনের ক্ষেত্রে একটি বিকল্প।

### 4. Kruskal-Wallis H test
*   **উদ্দেশ্য (Purpose):** তিন বা ততোধিক স্বাধীন গ্রুপের ডেটার ডিস্ট্রিবিউশন (distribution) বা মধ্যমা (median) তুলনা করা। এটি One-way ANOVA এর non-parametric বিকল্প。
*   **Assumptions (ধারণাসমূহ):**
    1.  **স্বাধীন পর্যবেক্ষণ (Independent observations):** সকল গ্রুপ স্বাধীন হতে হবে。
    2.  **পরিমাপের স্কেল (Scale of measurement):** ডিপেন্ডেন্ট ভেরিয়েবল অন্তত অর্ডিনাল (ordinal) স্কেলে পরিমাপযোগ্য হতে হবে。
    3.  **ডিস্ট্রিবিউশনের আকার (Shape of distribution):** যদি মধ্যমার তুলনা করতে চান, তবে সকল গ্রুপের ডিস্ট্রিবিউশনের আকার এবং বিস্তার (shape and spread) একই রকম হতে হবে。
*   **বিকল্প:** এটি নিজেই ANOVA এর নরমালিটি বা ভেরিয়েন্সের সমতা assumption লঙ্ঘনের ক্ষেত্রে একটি বিকল্প।

### 5. Fisher's exact test
*   **উদ্দেশ্য (Purpose):** দুটি ক্যাটেগরিক্যাল ভেরিয়েবলের মধ্যে অ্যাসোসিয়েশন (association) পরীক্ষা করা, বিশেষত যখন স্যাম্পল সাইজ ছোট হয় এবং 2x2 contingency table এ কিছু সেলের প্রত্যাশিত সংখ্যা (expected frequency) খুব কম থাকে (সাধারণত < 5)。
*   **Assumptions (ধারণাসমূহ):**
    1.  **স্বাধীন পর্যবেক্ষণ (Independent observations):** ডেটা পয়েন্টগুলো স্বাধীন হতে হবে。
    2.  **ক্যাটেগরিক্যাল ডেটা (Categorical data):** উভয় ভেরিয়েবল ক্যাটেগরিক্যাল হতে হবে。
    3.  **স্থির মার্জিন (Fixed marginals - কিছু ক্ষেত্রে):** টেস্টটি শর্তসাপেক্ষ, যা মার্জিনাল টোটালগুলোকে স্থির ধরে গণনা করে。
*   **বিকল্প:**
    *   যদি expected frequency যথেষ্ট বড় হয় (যেমন, সব সেলে > 5): **Chi-square test** ব্যবহার করা যেতে পারে।

### 6. Chi-square test (χ² test)
*   **উদ্দেশ্য (Purpose):**
    *   **Goodness-of-fit:** একটি ক্যাটেগরিক্যাল ভেরিয়েবলের পর্যবেক্ষণ করা ফ্রিকোয়েন্সি (observed frequency) একটি প্রত্যাশিত ফ্রিকোয়েন্সি ডিস্ট্রিবিউশনের (expected frequency distribution) সাথে মেলে কিনা তা পরীক্ষা করা。
    *   **Test of independence/association:** দুটি ক্যাটেগরিক্যাল ভেরিয়েবলের মধ্যে স্বাধীনতা বা অ্যাসোসিয়েশন পরীক্ষা করা (contingency table analysis)。
*   **Assumptions (Test of independence এর জন্য):**
    1.  **স্বাধীন পর্যবেক্ষণ (Independent observations):** ডেটা পয়েন্টগুলো স্বাধীন হতে হবে。
    2.  **ক্যাটেগরিক্যাল ডেটা (Categorical data):** উভয় ভেরিয়েবল ক্যাটেগরিক্যাল হতে হবে。
    3.  **স্যাম্পল সাইজ (Sample size):** স্যাম্পল সাইজ যথেষ্ট বড় হতে হবে। সাধারণত, contingency table এর 80% সেলের expected frequency ≥ 5 এবং কোনো সেলের expected frequency < 1 না হওয়া উচিত。
*   **যদি Assumptions লঙ্ঘিত হয়:**
    *   Expected frequency কম হলে (বিশেষত 2x2 টেবিলের জন্য): **Fisher's exact test**。
    *   বড় টেবিলের জন্য expected frequency কম হলে: সেল একত্রিত করা (cell merging), বা exact test এর অন্যান্য সংস্করণ ব্যবহার করা যেতে পারে।

### 7. Log-Rank test (Kaplan-Meier plot এর সাথে ব্যবহৃত)
*   **উদ্দেশ্য (Purpose):** দুটি বা ততোধিক গ্রুপের survival distributions (যেমন, একটি ঘটনা ঘটার সময় পর্যন্ত) তুলনা করা。
*   **Assumptions (ধারণাসমূহ):**
    1.  **স্বাধীন পর্যবেক্ষণ (Independent observations):** Survival times এবং censoring times স্বাধীন হতে হবে。
    2.  **নন-ইনফরমেটিভ সেন্সরিং (Non-informative censoring):** যে কারণে একজন অংশগ্রহণকারী সেন্সরড (censored) হয়, তা তার ঘটনা ঘটার সম্ভাবনার সাথে সম্পর্কিত নয়。
    3.  **প্রোপোরশনাল হ্যাজার্ডস (Proportional hazards):** সময়ের সাথে সাথে গ্রুপগুলোর মধ্যে hazard ratio (ঝুঁকির অনুপাত) ধ্রুবক থাকে। এর মানে হলো, একটি গ্রুপের তুলনায় অন্য গ্রুপের ঝুঁকি সময়ের সাথে পরিবর্তিত হয় না。
    4.  Survival times সঠিকভাবে সংজ্ঞায়িত এবং পরিমাপ করা হয়েছে。
*   **যদি Assumptions লঙ্ঘিত হয়:**
    *   Proportional hazards assumption লঙ্ঘিত হলে: **Cox model with time-dependent covariates**, অথবা অন্যান্য test যেমন **Wilcoxon (Gehan) test** (যা শুরুর দিকের পার্থক্যগুলিতে বেশি ওজন দেয়) বা **Tarone-Ware test** ব্যবহার করা যেতে পারে। গ্রাফিক্যাল অ্যাসেসমেন্ট (যেমন, log-minus-log plots) এই assumption পরীক্ষা করতে সাহায্য করে।

## Dependent (paired) Tests

এই পরীক্ষাগুলো ব্যবহৃত হয় যখন ডেটা পয়েন্টগুলো জোড়ায় জোড়ায় সম্পর্কিত থাকে, যেমন একই ব্যক্তির উপর বিভিন্ন সময়ে পরিমাপ বা ম্যাচড পেয়ার।

### 1. Paired t-test
*   **উদ্দেশ্য (Purpose):** একই সাবজেক্ট বা ম্যাচড পেয়ারের দুটি সম্পর্কিত মেজারমেন্টের গড় পার্থক্য (mean of differences) শূন্য কিনা তা পরীক্ষা করা。
*   **Assumptions (ধারণাসমূহ):**
    1.  **পেয়ারড পর্যবেক্ষণ (Paired observations):** ডেটা জোড়ায় আসতে হবে。
    2.  **স্বাধীন পার্থক্য (Independent differences):** পার্থক্যগুলো (differences between paired observations) একে অপরের থেকে স্বাধীন হতে হবে。
    3.  **পার্থক্যের নরমালিটি (Normality of differences):** পেয়ারড অবজারভেশনের পার্থক্যগুলো নরমালি ডিস্ট্রিবিউটেড হতে হবে। (ডেটা নিজেই নরমাল না হলেও চলবে, কিন্তু তাদের পার্থক্য নরমাল হতে হবে)。
    4.  **পরিমাপের স্কেল (Scale of measurement):** ডিপেন্ডেন্ট ভেরিয়েবল কন্টিনিউয়াস (continuous) হতে হবে。
*   **যদি Assumptions লঙ্ঘিত হয়:**
    *   পার্থক্যের নরমালিটি লঙ্ঘিত হলে: **Wilcoxon signed-rank test** (non-parametric বিকল্প)।

### 2. Repeated Measures ANOVA
*   **উদ্দেশ্য (Purpose):** একই সাবজেক্টের তিন বা ততোধিক সম্পর্কিত মেজারমেন্টের গড় (mean) তুলনা করা。
*   **Assumptions (ধারণাসমূহ):**
    1.  **স্বাধীন সাবজেক্ট (Independent subjects):** অংশগ্রহণকারীরা একে অপরের থেকে স্বাধীন。
    2.  **নরমালিটি (Normality):** প্রতিটি পরিমাপ সময়ে বা কন্ডিশনে ডিপেন্ডেন্ট ভেরিয়েবল নরমালি ডিস্ট্রিবিউটেড, অথবা মডেলের residuals নরমালি ডিস্ট্রিবিউটেড。
    3.  **Sphericity:** বিভিন্ন লেভেলের মধ্যে পার্থক্যের ভেরিয়েন্স (variances of the differences between levels) সমান হতে হবে। (Mauchly's test of sphericity দিয়ে পরীক্ষা করা যায়)। এটি compound symmetry এর একটি কম কঠোর রূপ。
    4.  **পরিমাপের স্কেল (Scale of measurement):** ডিপেন্ডেন্ট ভেরিয়েবল কন্টিনিউয়াস (continuous) হতে হবে。
*   **যদি Assumptions লঙ্ঘিত হয়:**
    *   Sphericity লঙ্ঘিত হলে: **Greenhouse-Geisser correction** বা **Huynh-Feldt correction** ব্যবহার করা হয় F-statistic এর degrees of freedom অ্যাডজাস্ট করার জন্য। অথবা **Multivariate ANOVA (MANOVA)** একটি বিকল্প হতে পারে কারণ এর জন্য sphericity assumption প্রয়োজন হয় না。
    *   নরমালিটি লঙ্ঘিত হলে: **Friedman test** (non-parametric বিকল্প)।

### 3. Mixed-effect Regression (Hierarchical Linear Models / Multilevel Models)
*   **উদ্দেশ্য (Purpose):** যখন ডেটার মধ্যে নেস্টেড বা হায়ারারকিক্যাল কাঠামো থাকে (যেমন, ছাত্ররা ক্লাসের মধ্যে, ক্লাস স্কুলের মধ্যে) অথবা যখন repeated measures থাকে এবং সময়ের সাথে সম্পর্ক জটিল হয় বা missing data থাকে। এটি fixed effects (গড় প্রতিক্রিয়া) এবং random effects (ব্যক্তি বা গ্রুপের মধ্যে তারতম্য) উভয়ই মডেল করতে পারে。
*   **Assumptions (ধারণাসমূহ):**
    1.  **লিনিয়ারিটি (Linearity):** প্রেডিক্টর এবং আউটকামের মধ্যে লিনিয়ার সম্পর্ক (প্রয়োজনে transformed)。
    2.  **স্বাধীন রেসিডুয়াল (Independence of residuals):** সর্বোচ্চ স্তরের হায়ারার্কিতে রেসিডুয়ালগুলো স্বাধীন হতে হবে (lower-level dependencies random effects দ্বারা মডেল করা হয়)。
    3.  **হোমোসিডাস্টিসিটি (Homoscedasticity of residuals):** রেসিডুয়ালের ভেরিয়েন্স ধ্রুবক হতে হবে。
    4.  **নরমালিটি (Normality of residuals):** মডেলের প্রতিটি স্তরে রেসিডুয়ালগুলো নরমালি ডিস্ট্রিবিউটেড হতে হবে。
    5.  **Random effects এর নরমালিটি:** Random effects গুলো নরমালি ডিস্ট্রিবিউটেড এবং তাদের গড় শূন্য。
*   **বিকল্প:** এটি একটি নমনীয় মডেলিং কৌশল এবং প্রায়শই অন্যান্য টেস্টের assumption লঙ্ঘিত হলে (বিশেষত ডেটার মধ্যে dependency থাকলে) বা আরও জটিল ডেটা স্ট্রাকচার মডেল করার জন্য ব্যবহৃত হয়। এর কোনো সরল "বিকল্প" নেই, কারণ এটি নিজেই একটি অ্যাডভান্সড সমাধান। মডেল স্পেসিফিকেশন এবং ডায়াগনস্টিকস এখানে অত্যন্ত গুরুত্বপূর্ণ।

### 4. Wilcoxon signed-rank test
*   **উদ্দেশ্য (Purpose):** একই সাবজেক্ট বা ম্যাচড পেয়ারের দুটি সম্পর্কিত মেজারমেন্টের ডিস্ট্রিবিউশনের (distribution) বা মধ্যমা পার্থক্যের (median of differences) তুলনা করা। এটি Paired t-test এর non-parametric বিকল্প, বিশেষত যখন পার্থক্যের নরমালিটির ধারণা লঙ্ঘিত হয়。
*   **Assumptions (ধারণাসমূহ):**
    1.  **পেয়ারড পর্যবেক্ষণ (Paired observations):** ডেটা জোড়ায় আসতে হবে。
    2.  **স্বাধীন পার্থক্য (Independent differences):** পার্থক্যগুলো একে অপরের থেকে স্বাধীন হতে হবে。
    3.  **সিমেট্রিক ডিস্ট্রিবিউশন (Symmetric distribution of differences):** পার্থক্যগুলো একটি সিমেট্রিক ডিস্ট্রিবিউশন থেকে আসে (মধ্যমার চারপাশে)。
    4.  **পরিমাপের স্কেল (Scale of measurement):** ডিপেন্ডেন্ট ভেরিয়েবল অন্তত অর্ডিনাল (ordinal) স্কেলে পরিমাপযোগ্য, তবে সাধারণত কন্টিনিউয়াস ডেটার জন্য ব্যবহৃত হয়。
*   **বিকল্প:** এটি নিজেই Paired t-test এর নরমালিটি assumption লঙ্ঘনের ক্ষেত্রে একটি বিকল্প।

### 5. Friedman test
*   **উদ্দেশ্য (Purpose):** একই সাবজেক্ট বা ব্লকের তিন বা ততোধিক সম্পর্কিত মেজারমেন্টের ডিস্ট্রিবিউশনের (distribution) তুলনা করা। এটি Repeated Measures ANOVA এর non-parametric বিকল্প。
*   **Assumptions (ধারণাসমূহ):**
    1.  **স্বাধীন ব্লক বা সাবজেক্ট (Independent blocks/subjects):** ডেটা ব্লক বা সাবজেক্ট দ্বারা গ্রুপ করা, যেখানে প্রতিটি ব্লকের মধ্যে পর্যবেক্ষণগুলো সম্পর্কিত কিন্তু বিভিন্ন ব্লকের পর্যবেক্ষণ স্বাধীন。
    2.  **কোনো ইন্টারঅ্যাকশন নেই (No interaction between blocks and treatments - কিছু ক্ষেত্রে):** সাধারণত ধরে নেওয়া হয় যে ব্লক এবং ট্রিটমেন্টের মধ্যে কোনো ইন্টারঅ্যাকশন নেই。
    3.  **পরিমাপের স্কেল (Scale of measurement):** ডিপেন্ডেন্ট ভেরিয়েবল অন্তত অর্ডিনাল (ordinal) স্কেলে পরিমাপযোগ্য。
    4.  অবজারভেশনগুলো প্রতিটি ব্লকের মধ্যে র‍্যাঙ্ক করা যায়。
*   **বিকল্প:** এটি নিজেই Repeated Measures ANOVA এর নরমালিটি assumption লঙ্ঘনের ক্ষেত্রে একটি বিকল্প। Post-hoc test (যেমন, Nemenyi test বা Conover test) ব্যবহার করা যেতে পারে নির্দিষ্ট গ্রুপগুলোর মধ্যে পার্থক্য খুঁজে বের করার জন্য।

### 6. McNemar's test
*   **উদ্দেশ্য (Purpose):** দুটি সম্পর্কিত বাইনারি ভেরিয়েবলের মধ্যে (যেমন, একটি ইন্টারভেনশনের আগে ও পরে একই সাবজেক্টের প্রতিক্রিয়া 'হ্যাঁ/না' বা 'সফল/ব্যর্থ') প্রান্তিক হোমোজিনিটি (marginal homogeneity) পরীক্ষা করা। এটি সাধারণত 2x2 contingency table এ paired categorical data এর জন্য ব্যবহৃত হয়, যেখানে আগ্রহের বিষয় হলো discordant pairs (যেখানে প্রতিক্রিয়া পরিবর্তিত হয়েছে)。
*   **Assumptions (ধারণাসমূহ):**
    1.  **পেয়ারড ডেটা (Paired nominal data):** ডেটা বাইনারি এবং পেয়ারড হতে হবে。
    2.  **স্বাধীন পেয়ার (Independent pairs):** পেয়ারগুলো একে অপরের থেকে স্বাধীন হতে হবে。
    3.  **স্যাম্পল সাইজ (Sample size):** Discordant pairs (যেখানে প্রথম পরিমাপে একটি ফলাফল এবং দ্বিতীয় পরিমাপে অন্য ফলাফল, অর্থাৎ b এবং c সেল) এর যোগফল যথেষ্ট বড় হওয়া উচিত (সাধারণত, b+c ≥ 10)。
*   **যদি Assumptions লঙ্ঘিত হয়:**
    *   Discordant pairs এর সংখ্যা কম হলে (b+c < 10): **Exact binomial test** ব্যবহার করা যেতে পারে McNemar's test এর একটি exact সংস্করণ হিসেবে。
    *   যদি তিন বা ততোধিক ক্যাটেগরি বা সময়বিন্দু থাকে: **Cochran's Q test** (McNemar's test এর এক্সটেনশন)।

# Statistical Tests: শূন্য থেকে শুরু - A Complete Beginner's Guide

## পরিসংখ্যানিক পরীক্ষা কী এবং কেন প্রয়োজন?

### মূল ধারণা (Basic Concept)
পরিসংখ্যানিক পরীক্ষা (Statistical Tests) হলো এমন একটি পদ্ধতি যা দিয়ে আমরা ডেটার মাধ্যমে কোনো hypothesis (অনুমান) সত্য কি মিথ্যা তা যাচাই করি। এটি একটি গাণিতিক পদ্ধতি যা আমাদের বলে দেয় যে আমাদের পর্যবেক্ষণ করা পার্থক্য সত্যিই গুরুত্বপূর্ণ নাকি কেবলমাত্র সাংযোগিক (random chance)।

### Real-life উদাহরণ:
**সাধারণ জীবনের উদাহরণ:** ধরুন আপনি জানতে চান যে পুরুষ এবং মহিলাদের গড় উচ্চতার মধ্যে পার্থক্য আছে কিনা। আপনি ১০০ জন পুরুষ এবং ১০০ জন মহিলার উচ্চতা পরিমাপ করলেন। দেখা গেল পুরুষদের গড় উচ্চতা ৫.৭ ফুট এবং মহিলাদের ৫.৩ ফুট। এখন প্রশ্ন হলো: এই পার্থক্যটি সত্যিই গুরুত্বপূর্ণ নাকি কেবল coincidence?

**চিকিৎসা বিজ্ঞানের উদাহরণ:** একটি নতুন ওষুধ কতটা কার্যকর তা পরীক্ষা করতে হবে। ১০০ জন রোগীকে নতুন ওষুধ দেওয়া হলো এবং আরও ১০০ জনকে পুরানো ওষুধ দেওয়া হলো। নতুন ওষুধ গ্রুপে ৮০% রোগী সুস্থ হলো, পুরানো ওষুধ গ্রুপে ৭০% সুস্থ হলো। এই পার্থক্য significant কিনা?

### মৌলিক Q&A:

**প্রশ্ন ১:** Statistical test কেন করি?
**উত্তর:** কারণ আমরা জানতে চাই আমাদের observation সত্যিই meaningful নাকি random। উদাহরণ: মুদ্রা ১০ বার ছুড়ে ৭ বার চিৎ এলে, এটি কি coincidence নাকি মুদ্রাটি biased?

**প্রশ্ন ২:** P-value কী?
**উত্তর:** P-value হলো সম্ভাবনা যে null hypothesis সত্য হলে আপনার observed result বা তার চেয়ে extreme result পাওয়ার সম্ভাবনা কত। সাধারণত p < 0.05 হলে আমরা বলি result significant।

**প্রশ্ন ৩:** Null hypothesis কী?
**উত্তর:** Null hypothesis (H₀) হলো আমাদের "default" assumption। সাধারণত এটি বলে যে "কোনো পার্থক্য নেই" বা "কোনো effect নেই"। Alternative hypothesis (H₁) বলে যে "পার্থক্য আছে"।

# Part 1: Independent (Un-paired) Tests - বিস্তারিত টিউটোরিয়াল

## 1. Student's t-test - দুটি গ্রুপের গড় তুলনা

### সাধারণ ধারণা:
Student's t-test ব্যবহার করা হয় দুটি স্বাধীন গ্রুপের গড় (mean) তুলনা করার জন্য। এটি বলে দেয় যে দুটি গ্রুপের গড়ের মধ্যে যে পার্থক্য আমরা দেখছি, সেটা সত্যিই গুরুত্বপূর্ণ নাকি কেবল সাংযোগিক।

### কখন ব্যবহার করবেন:
- দুটি আলাদা গ্রুপ আছে (যেমন: পুরুষ-মহিলা, চিকিৎসা-নিয়ন্ত্রণ)
- তুলনা করতে চান একটি continuous variable (যেমন: উচ্চতা, ওজন, স্কোর)
- ডেটা approximately নরমাল distribution অনুসরণ করে

### Real-life উদাহরণ:
**শিক্ষা:** দুটি ভিন্ন শিক্ষা পদ্ধতি (traditional vs online) এর পরীক্ষার স্কোর তুলনা করা।
**স্বাস্থ্য:** নতুন ব্যায়াম পদ্ধতি vs পুরানো পদ্ধতির weight loss তুলনা করা।
**ব্যবসা:** দুটি ভিন্ন marketing strategy এর sales তুলনা করা।

### Q&A সেশন:

**প্রশ্ন ১:** কেন "Student's" বলা হয়?
**উত্তর:** William Gosset নামের একজন statistician এর ছদ্মনাম ছিল "Student"। তিনি Guinness brewery তে কাজ করতেন এবং ছোট sample size এর জন্য এই test তৈরি করেছিলেন।

**প্রশ্ন ২:** One-tailed vs Two-tailed test এর পার্থক্য কী?
**উত্তর:** 
- Two-tailed: "দুটি গ্রুপের গড় আলাদা" (কোনটা বড়/ছোট বলা হয় না)
- One-tailed: "গ্রুপ A এর গড় গ্রুপ B এর চেয়ে বড়" (নির্দিষ্ট দিক বলা হয়)

**প্রশ্ন ৩:** Sample size কত হওয়া উচিত?
**উত্তর:** প্রতি গ্রুপে অন্তত ৩০ জন থাকলে ভালো। কম হলেও চলে, কিন্তু normality assumption আরো গুরুত্বপূর্ণ হয়।

### R কোড এবং বিস্তারিত ব্যাখ্যা:

```r
# Student's t-test - Practical R Example
# আমরা দুটি গ্রুপের ডেটা তৈরি করব

# প্রয়োজনীয় লাইব্রেরি লোড করি
library(ggplot2)  # গ্রাফ তৈরির জন্য
library(dplyr)    # ডেটা manipulation এর জন্য

# Step 1: Simulated Data তৈরি করি
# ধরি, দুটি ভিন্ন coaching center এর students এর পরীক্ষার স্কোর
set.seed(123)  # reproducible results এর জন্য

# Group A: Traditional coaching (গড় স্কোর ৭৫, standard deviation ১০)
group_A <- rnorm(n = 50, mean = 75, sd = 10)

# Group B: Online coaching (গড় স্কোর ৮০, standard deviation ১২)  
group_B <- rnorm(n = 45, mean = 80, sd = 12)

# Step 2: ডেটা দেখি
cat("Group A (Traditional) - Mean:", round(mean(group_A), 2), ", SD:", round(sd(group_A), 2), "\n")
cat("Group B (Online) - Mean:", round(mean(group_B), 2), ", SD:", round(sd(group_B), 2), "\n")

# Step 3: Data visualization
# দুটি গ্রুপের ডিস্ট্রিবিউশন দেখি
data_combined <- data.frame(
  scores = c(group_A, group_B),
  group = c(rep("Traditional", length(group_A)), rep("Online", length(group_B)))
)

# Histogram
ggplot(data_combined, aes(x = scores, fill = group)) +
  geom_histogram(alpha = 0.7, position = "identity", bins = 15) +
  labs(title = "Distribution of Test Scores by Coaching Method",
       x = "Test Scores", y = "Frequency") +
  theme_minimal()

# Box plot
ggplot(data_combined, aes(x = group, y = scores, fill = group)) +
  geom_boxplot(alpha = 0.7) +
  geom_jitter(width = 0.2, alpha = 0.5) +
  labs(title = "Comparison of Test Scores",
       x = "Coaching Method", y = "Test Scores") +
  theme_minimal()

# Step 4: Assumptions Check
# Assumption 1: Normality check
# Shapiro-Wilk test (n < 50 এর জন্য ভালো)
shapiro_A <- shapiro.test(group_A)
shapiro_B <- shapiro.test(group_B)

cat("Normality Tests:\n")
cat("Group A p-value:", round(shapiro_A$p.value, 4), 
    ifelse(shapiro_A$p.value > 0.05, "(Normal)", "(Not Normal)"), "\n")
cat("Group B p-value:", round(shapiro_B$p.value, 4), 
    ifelse(shapiro_B$p.value > 0.05, "(Normal)", "(Not Normal)"), "\n")

# Q-Q plots for visual check
par(mfrow = c(1, 2))
qqnorm(group_A, main = "Q-Q Plot: Traditional")
qqline(group_A)
qqnorm(group_B, main = "Q-Q Plot: Online")
qqline(group_B)
par(mfrow = c(1, 1))

# Assumption 2: Equal variances check (Levene's Test)
# F-test for equality of variances
f_test <- var.test(group_A, group_B)
cat("Equal Variances Test (F-test):\n")
cat("F-statistic:", round(f_test$statistic, 4), ", p-value:", round(f_test$p.value, 4))
cat(ifelse(f_test$p.value > 0.05, " (Equal variances)", " (Unequal variances)"), "\n")

# Step 5: Perform t-test
# যদি assumptions meet করে তাহলে standard t-test
# যদি variances unequal থাকে তাহলে Welch's t-test

if(f_test$p.value > 0.05) {
  # Equal variances assumed
  t_result <- t.test(group_A, group_B, var.equal = TRUE)
  cat("\nStudent's t-test (Equal variances assumed):\n")
} else {
  # Unequal variances (Welch's t-test)
  t_result <- t.test(group_A, group_B, var.equal = FALSE)
  cat("\nWelch's t-test (Unequal variances):\n")
}

print(t_result)

# Step 6: Results Interpretation
cat("\n--- Results Interpretation ---\n")
cat("Mean difference:", round(t_result$estimate[1] - t_result$estimate[2], 2), "\n")
cat("95% Confidence Interval:", round(t_result$conf.int[1], 2), "to", round(t_result$conf.int[2], 2), "\n")
cat("t-statistic:", round(t_result$statistic, 4), "\n")
cat("degrees of freedom:", round(t_result$parameter, 2), "\n")
cat("p-value:", round(t_result$p.value, 6), "\n")

if(t_result$p.value < 0.05) {
  cat("Conclusion: Significant difference exists between the two groups (p < 0.05)\n")
  cat("We reject the null hypothesis.\n")
} else {
  cat("Conclusion: No significant difference between the two groups (p >= 0.05)\n")
  cat("We fail to reject the null hypothesis.\n")
}

# Effect size calculation (Cohen's d)
pooled_sd <- sqrt(((length(group_A)-1)*var(group_A) + (length(group_B)-1)*var(group_B)) / 
                 (length(group_A) + length(group_B) - 2))
cohens_d <- (mean(group_A) - mean(group_B)) / pooled_sd
cat("Cohen's d (Effect size):", round(cohens_d, 3))

if(abs(cohens_d) < 0.2) {
  cat(" (Small effect)\n")
} else if(abs(cohens_d) < 0.5) {
  cat(" (Small to medium effect)\n")
} else if(abs(cohens_d) < 0.8) {
  cat(" (Medium to large effect)\n")
} else {
  cat(" (Large effect)\n")
}
```

### R কোড ব্যাখ্যা:

**1. Data Generation:**
- `rnorm()` function দিয়ে normal distribution থেকে random data তৈরি করি
- `set.seed()` দিয়ে reproducible results নিশ্চিত করি

**2. Visualization:**
- Histogram দিয়ে distribution এর shape দেখি
- Box plot দিয়ে median, quartiles এবং outliers দেখি

**3. Assumption Testing:**
- **Normality:** Shapiro-Wilk test এবং Q-Q plot
- **Equal Variances:** F-test বা Levene's test

**4. Test Selection:**
- Equal variances: Student's t-test (`var.equal = TRUE`)
- Unequal variances: Welch's t-test (`var.equal = FALSE`)

**5. Effect Size:**
- Cohen's d calculation করি practical significance বোঝার জন্য

### Advanced Q&A:

**প্রশ্ন ৪:** Cohen's d কী এবং কেন গুরুত্বপূর্ণ?
**উত্তর:** Effect size measure। Statistical significance (p-value) বলে পার্থক্য আছে কিনা, কিন্তু Cohen's d বলে সেই পার্থক্য কতটা meaningful। 
- 0.2 = Small effect
- 0.5 = Medium effect  
- 0.8 = Large effect

**প্রশ্ন ৫:** Welch's t-test কখন ব্যবহার করব?
**উত্তর:** যখন দুটি গ্রুপের variance significantly different হয়। এটি Student's t-test এর চেয়ে robust এবং আজকাল default choice।

## 2. Mann-Whitney U Test - Non-parametric Alternative

### সাধারণ ধারণা:
Mann-Whitney U test (অন্য নাম: Wilcoxon rank-sum test) হলো t-test এর non-parametric বিকল্প। এটি ব্যবহার করা হয় যখন:
- ডেটা নরমাল ডিস্ট্রিবিউশন অনুসরণ করে না
- Sample size খুব ছোট
- Ordinal data আছে

### কখন ব্যবহার করবেন:
- Normality assumption লঙ্ঘিত হলে
- Extreme outliers থাকলে
- Ordinal scale data (যেমন: rating, ranking)
- Small sample sizes

### Real-life উদাহরণ:
**চিকিৎসা:** দুটি চিকিৎসা পদ্ধতির pain relief rating (1-10 scale) তুলনা
**শিক্ষা:** দুটি বিভাগের students এর satisfaction survey (Likert scale) তুলনা
**ব্যবসা:** দুটি product এর customer rating তুলনা

### কিভাবে কাজ করে:
1. সব ডেটা একসাথে নিয়ে rank করে (ছোট থেকে বড়)
2. প্রতি গ্রুপের rank এর sum বের করে
3. Expected rank sum এর সাথে তুলনা করে
4. Significant difference আছে কিনা দেখে

### Q&A সেশন:

**প্রশ্ন ১:** কেন "U" বলা হয়?
**উত্তর:** Test statistic টির নাম "U"। এটি একটি গ্রুপের observations কতগুলো অন্য গ্রুপের observations এর চেয়ে বড় তা count করে।

**প্রশ্ন ২:** Mann-Whitney vs t-test: কোনটা ভালো?
**উত্তর:** 
- **t-test:** আরো powerful (যদি assumptions meet করে)
- **Mann-Whitney:** আরো robust (assumptions এর ব্যাপারে কম strict)

**প্রশ্ন ৩:** Tied values থাকলে কী হবে?
**উত্তর:** R automatically correction apply করে tied values এর জন্য।

### R কোড এবং বিস্তারিত ব্যাখ্যা:

```r
# Mann-Whitney U Test - Practical R Example
# ধরি, দুটি hospital এর patient satisfaction score (1-10 scale)

# Step 1: Non-normal data তৈরি করি
set.seed(456)

# Hospital A: Skewed distribution (বেশিরভাগ রোগী মোটামুটি সন্তুষ্ট)
hospital_A <- c(
  rep(6, 8), rep(7, 12), rep(8, 15), rep(9, 10), rep(10, 5)  # Positively skewed
)

# Hospital B: Different distribution (আরো varied responses)
hospital_B <- c(
  rep(4, 3), rep(5, 7), rep(6, 8), rep(7, 8), rep(8, 6), rep(9, 8)  # More spread out
)

# Add some random noise
hospital_A <- hospital_A + runif(length(hospital_A), -0.3, 0.3)
hospital_B <- hospital_B + runif(length(hospital_B), -0.3, 0.3)

# Step 2: Basic descriptive statistics
cat("Hospital A - Median:", median(hospital_A), ", IQR:", IQR(hospital_A), "\n")
cat("Hospital B - Median:", median(hospital_B), ", IQR:", IQR(hospital_B), "\n")

# Step 3: Visualization
data_satisfaction <- data.frame(
  scores = c(hospital_A, hospital_B),
  hospital = c(rep("Hospital A", length(hospital_A)), rep("Hospital B", length(hospital_B)))
)

# Histogram to see the distribution
ggplot(data_satisfaction, aes(x = scores, fill = hospital)) +
  geom_histogram(alpha = 0.7, position = "identity", bins = 10) +
  labs(title = "Patient Satisfaction Scores Distribution",
       x = "Satisfaction Score (1-10)", y = "Count") +
  theme_minimal() +
  facet_wrap(~hospital, ncol = 1)

# Box plot for comparison
ggplot(data_satisfaction, aes(x = hospital, y = scores, fill = hospital)) +
  geom_boxplot(alpha = 0.7) +
  geom_jitter(width = 0.2, alpha = 0.6) +
  labs(title = "Patient Satisfaction Comparison",
       x = "Hospital", y = "Satisfaction Score") +
  theme_minimal()

# Step 4: Check normality assumption
# যদি এটি violated হয় তাহলে Mann-Whitney ব্যবহার করব
shapiro_A_sat <- shapiro.test(hospital_A)
shapiro_B_sat <- shapiro.test(hospital_B)

cat("\nNormality Tests:\n")
cat("Hospital A p-value:", round(shapiro_A_sat$p.value, 4), 
    ifelse(shapiro_A_sat$p.value > 0.05, "(Normal)", "(Not Normal)"), "\n")
cat("Hospital B p-value:", round(shapiro_B_sat$p.value, 4), 
    ifelse(shapiro_B_sat$p.value > 0.05, "(Normal)", "(Not Normal)"), "\n")

# Step 5: Perform Mann-Whitney U test
mannwhitney_result <- wilcox.test(hospital_A, hospital_B, alternative = "two.sided")

cat("\n--- Mann-Whitney U Test Results ---\n")
print(mannwhitney_result)

# Step 6: Additional information
# Manual calculation of U statistic for understanding
n1 <- length(hospital_A)
n2 <- length(hospital_B)
R1 <- sum(rank(c(hospital_A, hospital_B))[1:n1])  # Sum of ranks for group 1
U1 <- R1 - n1*(n1+1)/2  # U statistic for group 1
U2 <- n1*n2 - U1        # U statistic for group 2

cat("\nManual Calculation:\n")
cat("Sample sizes: n1 =", n1, ", n2 =", n2, "\n")
cat("Sum of ranks for Hospital A:", R1, "\n")
cat("U statistic for Hospital A:", U1, "\n")
cat("U statistic for Hospital B:", U2, "\n")
cat("Test uses smaller U:", min(U1, U2), "\n")

# Step 7: Effect size (r = Z/sqrt(N))
# Convert W to Z-score approximation for large samples
W <- mannwhitney_result$statistic
mu_W <- n1 * n2 / 2
sigma_W <- sqrt(n1 * n2 * (n1 + n2 + 1) / 12)
Z <- (W - mu_W) / sigma_W
effect_size_r <- abs(Z) / sqrt(n1 + n2)

cat("Effect size (r):", round(effect_size_r, 3))
if(effect_size_r < 0.1) {
  cat(" (Negligible effect)\n")
} else if(effect_size_r < 0.3) {
  cat(" (Small effect)\n")
} else if(effect_size_r < 0.5) {
  cat(" (Medium effect)\n")
} else {
  cat(" (Large effect)\n")
}

# Step 8: Results interpretation
cat("\n--- Results Interpretation ---\n")
cat("Median Hospital A:", round(median(hospital_A), 2), "\n")
cat("Median Hospital B:", round(median(hospital_B), 2), "\n")
cat("p-value:", round(mannwhitney_result$p.value, 6), "\n")

if(mannwhitney_result$p.value < 0.05) {
  cat("Conclusion: Significant difference in satisfaction scores between hospitals (p < 0.05)\n")
  if(median(hospital_A) > median(hospital_B)) {
    cat("Hospital A has significantly higher satisfaction scores.\n")
  } else {
    cat("Hospital B has significantly higher satisfaction scores.\n")
  }
} else {
  cat("Conclusion: No significant difference in satisfaction scores (p >= 0.05)\n")
}

# Step 9: Compare with t-test (just for educational purpose)
cat("\n--- Comparison with t-test ---\n")
t_test_comparison <- t.test(hospital_A, hospital_B)
cat("t-test p-value:", round(t_test_comparison$p.value, 6), "\n")
cat("Mann-Whitney p-value:", round(mannwhitney_result$p.value, 6), "\n")

if(abs(t_test_comparison$p.value - mannwhitney_result$p.value) < 0.01) {
  cat("Both tests give similar results.\n")
} else {
  cat("Tests give different results - Mann-Whitney is more appropriate here.\n")
}
```

## 3. One-way ANOVA - তিন বা অধিক গ্রুপের তুলনা

### সাধারণ ধারণা:
ANOVA (Analysis of Variance) ব্যবহার করা হয় যখন তিন বা ততোধিক গ্রুপের গড় তুলনা করতে হয়। এটি t-test এর extension। যদি আমরা multiple t-tests করি তাহলে Type I error বেড়ে যায়, তাই ANOVA ব্যবহার করি।

### কখন ব্যবহার করবেন:
- তিন বা ততোধিক স্বাধীন গ্রুপ
- একটি continuous dependent variable
- গ্রুপগুলোর গড় তুলনা করতে চান

### Real-life উদাহরণ:
**শিক্ষা:** তিনটি ভিন্ন teaching method এর effectiveness তুলনা
**চিকিৎসা:** চারটি ভিন্ন drug dosage এর effectiveness তুলনা  
**ব্যবসা:** পাঁচটি ভিন্ন marketing strategy এর sales performance তুলনা

### Q&A সেশন:

**প্রশ্ন ১:** কেন multiple t-tests না করে ANOVA করি?
**উত্তর:** Multiple comparisons করলে Type I error rate বেড়ে যায়। যেমন, ৩টি গ্রুপের জন্য ৩টি t-test করলে overall error rate হবে 1-(0.95)³ = 14.3%!

**প্রশ্ন ২:** F-statistic কী?
**উত্তর:** F = Between-group variance / Within-group variance। যদি গ্রুপগুলোর মধ্যে সত্যিই পার্থক্য থাকে তাহলে F বড় হবে।

**প্রশ্ন ৩:** Post-hoc test কী?
**উত্তর:** ANOVA significant হলে এটি বলে কোথাও পার্থক্য আছে, কিন্তু কোন গ্রুপগুলোর মধ্যে তা বলে না। সেটা জানার জন্য post-hoc tests (যেমন Tukey HSD) করি।

```r
# One-way ANOVA - Practical R Example
# তিনটি ভিন্ন diet plan এর weight loss effectiveness তুলনা

# প্রয়োজনীয় লাইব্রেরি
library(ggplot2)    # visualization এর জন্য
library(dplyr)      # data manipulation এর জন্য  
library(car)        # Levene's test এর জন্য
library(agricolae)  # post-hoc tests এর জন্য

# Step 1: Simulated data তৈরি করি (তিনটি diet plan)
set.seed(789)

# Diet A: Low-carb (গড় weight loss ৮ kg, sd ২.৫)
diet_A <- rnorm(n = 25, mean = 8, sd = 2.5)

# Diet B: Mediterranean (গড় weight loss ৬ kg, sd ৩)  
diet_B <- rnorm(n = 30, mean = 6, sd = 3)

# Diet C: Intermittent Fasting (গড় weight loss ১০ kg, sd ২)
diet_C <- rnorm(n = 28, mean = 10, sd = 2)

# Data frame তৈরি করি
diet_data <- data.frame(
  weight_loss = c(diet_A, diet_B, diet_C),
  diet_type = factor(c(rep("Low-carb", length(diet_A)), 
                      rep("Mediterranean", length(diet_B)),
                      rep("Intermittent", length(diet_C))))
)

# Step 2: Descriptive statistics দেখি
cat("=== Descriptive Statistics ===\n")
summary_stats <- diet_data %>%
  group_by(diet_type) %>%
  summarise(
    n = n(),                           # sample size
    mean = round(mean(weight_loss), 2), # গড়
    sd = round(sd(weight_loss), 2),     # standard deviation
    median = round(median(weight_loss), 2), # মধ্যমা
    min = round(min(weight_loss), 2),   # সর্বনিম্ন
    max = round(max(weight_loss), 2)    # সর্বোচ্চ
  )
print(summary_stats)

# Step 3: Data visualization
# Box plot - গ্রুপ comparison এর জন্য best
ggplot(diet_data, aes(x = diet_type, y = weight_loss, fill = diet_type)) +
  geom_boxplot(alpha = 0.7) +                    # box plot
  geom_jitter(width = 0.2, alpha = 0.5) +       # individual points দেখানোর জন্য
  labs(title = "Weight Loss by Diet Type",
       x = "Diet Type", 
       y = "Weight Loss (kg)",
       fill = "Diet Type") +
  theme_minimal() +
  theme(legend.position = "none")        # legend remove করি x-axis এ আছে বলে

# Histogram faceted by group - distribution shape দেখার জন্য
ggplot(diet_data, aes(x = weight_loss, fill = diet_type)) +
  geom_histogram(bins = 12, alpha = 0.7, color = "black") +
  facet_wrap(~diet_type, ncol = 1) +     # আলাদা আলাদা panel এ দেখাই
  labs(title = "Distribution of Weight Loss by Diet Type",
       x = "Weight Loss (kg)", 
       y = "Frequency") +
  theme_minimal()

# Step 4: ANOVA Assumptions Check
cat("\n=== ANOVA Assumptions Check ===\n")

# Assumption 1: Normality check (প্রতি গ্রুপে)
# Shapiro-Wilk test প্রতি গ্রুপের জন্য
normality_results <- diet_data %>%
  group_by(diet_type) %>%
  summarise(
    shapiro_p = round(shapiro.test(weight_loss)$p.value, 4),
    is_normal = ifelse(shapiro_p > 0.05, "Yes", "No")
  )

cat("Normality Test Results:\n")
print(normality_results)

# Visual normality check - Q-Q plots
par(mfrow = c(1, 3))  # ১ row, ৩ columns এ arrange করি
for(diet in levels(diet_data$diet_type)) {
  data_subset <- diet_data$weight_loss[diet_data$diet_type == diet]
  qqnorm(data_subset, main = paste("Q-Q Plot:", diet))
  qqline(data_subset, col = "red")  # reference line
}
par(mfrow = c(1, 1))  # reset layout

# Assumption 2: Homogeneity of variances (Levene's test)
levene_result <- leveneTest(weight_loss ~ diet_type, data = diet_data)
cat("\nLevene's Test for Equal Variances:\n")
cat("F-statistic:", round(levene_result$`F value`[1], 4), "\n")
cat("p-value:", round(levene_result$`Pr(>F)`[1], 4), "\n")
cat("Equal variances:", ifelse(levene_result$`Pr(>F)`[1] > 0.05, "Yes", "No"), "\n")

# Step 5: Perform One-way ANOVA
cat("\n=== One-way ANOVA Results ===\n")

# ANOVA model fit করি
anova_model <- aov(weight_loss ~ diet_type, data = diet_data)

# ANOVA table দেখি
anova_summary <- summary(anova_model)
print(anova_summary)

# Effect size calculation (Eta-squared)
# SS_between / SS_total
ss_between <- anova_summary[[1]]["diet_type", "Sum Sq"]
ss_total <- sum(anova_summary[[1]]["Sum Sq"])
eta_squared <- ss_between / ss_total

cat("\nEffect Size (Eta-squared):", round(eta_squared, 3))
if(eta_squared < 0.01) {
  cat(" (Small effect)\n")
} else if(eta_squared < 0.06) {
  cat(" (Medium effect)\n") 
} else if(eta_squared < 0.14) {
  cat(" (Large effect)\n")
} else {
  cat(" (Very large effect)\n")
}

# Step 6: Post-hoc tests (যদি ANOVA significant হয়)
anova_p_value <- anova_summary[[1]]["diet_type", "Pr(>F)"]

if(anova_p_value < 0.05) {
  cat("\n=== Post-hoc Tests (Tukey HSD) ===\n")
  cat("ANOVA is significant, conducting pairwise comparisons...\n")
  
  # Tukey HSD test - সব possible pairs compare করে
  tukey_results <- TukeyHSD(anova_model)
  print(tukey_results)
  
  # Plot Tukey HSD results
  plot(tukey_results, las = 1)  # las=1 makes y-axis labels horizontal
  
  # Alternative: Using agricolae package for better formatting
  tukey_agricolae <- HSD.test(anova_model, "diet_type", alpha = 0.05)
  cat("\nTukey HSD Groups (Similar letters = No significant difference):\n")
  print(tukey_agricolae$groups)
  
} else {
  cat("\nANOVA is not significant (p >= 0.05)")
  cat("\nNo need for post-hoc tests\n")
}

# Step 7: Model diagnostics
cat("\n=== Model Diagnostics ===\n")

# Residuals analysis
residuals <- residuals(anova_model)
fitted_values <- fitted(anova_model)

# 1. Residuals vs Fitted (homoscedasticity check)
par(mfrow = c(2, 2))
plot(fitted_values, residuals, 
     xlab = "Fitted Values", ylab = "Residuals",
     main = "Residuals vs Fitted")
abline(h = 0, col = "red", lty = 2)  # horizontal line at 0

# 2. Normal Q-Q plot of residuals
qqnorm(residuals, main = "Normal Q-Q Plot of Residuals")
qqline(residuals, col = "red")

# 3. Scale-Location plot  
plot(fitted_values, sqrt(abs(residuals)),
     xlab = "Fitted Values", ylab = "√|Residuals|",
     main = "Scale-Location Plot")

# 4. Residuals vs Leverage
plot(anova_model, which = 5)  # Cook's distance plot
par(mfrow = c(1, 1))

# Step 8: Results interpretation এবং conclusion
cat("\n=== Results Summary ===\n")
cat("F-statistic:", round(anova_summary[[1]]["diet_type", "F value"], 4), "\n")
cat("p-value:", round(anova_p_value, 6), "\n")
cat("Degrees of freedom:", anova_summary[[1]]["diet_type", "Df"], 
    "and", anova_summary[[1]]["Residuals", "Df"], "\n")

if(anova_p_value < 0.05) {
  cat("\nConclusion: Significant difference exists among diet types (p < 0.05)\n")
  cat("We reject the null hypothesis.\n")
  cat("At least one diet type has significantly different weight loss.\n")
} else {
  cat("\nConclusion: No significant difference among diet types (p >= 0.05)\n") 
  cat("We fail to reject the null hypothesis.\n")
  cat("All diet types show similar weight loss effectiveness.\n")
}

# Alternative: Welch ANOVA (যদি equal variances assumption violated হয়)
if(levene_result$`Pr(>F)`[1] < 0.05) {
  cat("\n=== Welch ANOVA (Unequal Variances) ===\n")
  welch_anova <- oneway.test(weight_loss ~ diet_type, data = diet_data)
  print(welch_anova)
}
```

## 4. Kruskal-Wallis H Test - ANOVA এর Non-parametric বিকল্প

### সাধারণ ধারণা:
Kruskal-Wallis test হলো One-way ANOVA এর non-parametric version। এটি ব্যবহার করা হয় যখন ANOVA এর assumptions (বিশেষত normality) লঙ্ঘিত হয়। এটি ranks ব্যবহার করে কাজ করে।

### কখন ব্যবহার করবেন:
- তিন বা ততোধিক স্বাধীন গ্রুপ
- Normality assumption violated
- Ordinal data বা non-normal continuous data
- Extreme outliers থাকলে

### Real-life উদাহরণ:
**শিক্ষা:** তিনটি বিভিন্ন university এর student satisfaction rating (1-5 scale)
**চিকিৎসা:** চারটি ভিন্ন treatment এর pain relief rating
**ব্যবসা:** পাঁচটি product এর customer preference ranking

### Q&A সেশন:

**প্রশ্ন ১:** Kruskal-Wallis কিভাবে কাজ করে?
**উত্তর:** সব observations একসাথে rank করে, তারপর প্রতি গ্রুপের rank sum calculate করে। যদি groups সত্যিই different হয় তাহলে rank sums significantly different হবে।

**প্রশ্ন ২:** Post-hoc test কোনটা ব্যবহার করব?
**উত্তর:** Dunn's test সবচেয়ে popular। এটি pairwise comparisons করে with appropriate correction।

**প্রশ্ন ৩:** ANOVA vs Kruskal-Wallis: কোনটা বেশি powerful?
**উত্তর:** ANOVA বেশি powerful যদি assumptions meet করে। কিন্তু Kruskal-Wallis বেশি robust।

```r
# Kruskal-Wallis H Test - Practical R Example  
# তিনটি ভিন্ন treatment এর patient pain relief rating (1-10 scale)

# প্রয়োজনীয় libraries
library(ggplot2)
library(dplyr) 
library(dunn.test)  # post-hoc tests এর জন্য
library(FSA)        # alternative post-hoc tests

# Step 1: Non-normal, ordinal-like data তৈরি করি
set.seed(321)

# Treatment A: Traditional medicine (mostly moderate relief, skewed)
treat_A <- c(rep(3, 5), rep(4, 8), rep(5, 12), rep(6, 10), rep(7, 5))

# Treatment B: New drug (more varied, some very high relief)  
treat_B <- c(rep(2, 3), rep(4, 4), rep(6, 8), rep(8, 10), rep(9, 7), rep(10, 3))

# Treatment C: Physiotherapy (consistently moderate)
treat_C <- c(rep(4, 8), rep(5, 15), rep(6, 12), rep(7, 5))

# Add some small random noise to make it more realistic
treat_A <- treat_A + runif(length(treat_A), -0.4, 0.4)
treat_B <- treat_B + runif(length(treat_B), -0.4, 0.4)  
treat_C <- treat_C + runif(length(treat_C), -0.4, 0.4)

# Ensure values stay within 1-10 range
treat_A <- pmax(1, pmin(10, treat_A))
treat_B <- pmax(1, pmin(10, treat_B))
treat_C <- pmax(1, pmin(10, treat_C))

# Create data frame
pain_data <- data.frame(
  relief_score = c(treat_A, treat_B, treat_C),
  treatment = factor(c(rep("Traditional", length(treat_A)),
                      rep("New_Drug", length(treat_B)), 
                      rep("Physiotherapy", length(treat_C))))
)

# Step 2: Descriptive statistics 
cat("=== Descriptive Statistics ===\n")
desc_stats <- pain_data %>%
  group_by(treatment) %>%
  summarise(
    n = n(),                                    # sample size
    median = round(median(relief_score), 2),    # median (center measure for non-normal)
    IQR = round(IQR(relief_score), 2),         # inter-quartile range
    mean = round(mean(relief_score), 2),        # mean (for comparison)
    sd = round(sd(relief_score), 2),           # standard deviation
    min = round(min(relief_score), 2),
    max = round(max(relief_score), 2)
  )
print(desc_stats)

# Step 3: Data visualization
# Box plot - median and quartiles দেখার জন্য perfect
ggplot(pain_data, aes(x = treatment, y = relief_score, fill = treatment)) +
  geom_boxplot(alpha = 0.7, outlier.shape = 16) +        # outliers show করি
  geom_jitter(width = 0.2, alpha = 0.4) +               # individual points
  labs(title = "Pain Relief Scores by Treatment Type",
       x = "Treatment", 
       y = "Pain Relief Score (1-10)",
       fill = "Treatment") +
  theme_minimal() +
  theme(legend.position = "none")

# Violin plot - distribution shape better দেখার জন্য  
ggplot(pain_data, aes(x = treatment, y = relief_score, fill = treatment)) +
  geom_violin(alpha = 0.7) +                            # distribution shape
  geom_boxplot(width = 0.1, alpha = 0.8) +             # box plot overlay
  labs(title = "Distribution of Pain Relief Scores",
       x = "Treatment", y = "Pain Relief Score") +
  theme_minimal()

# Step 4: Check ANOVA assumptions (to justify using Kruskal-Wallis)
cat("\n=== Checking ANOVA Assumptions ===\n")

# Normality test for each group
normality_check <- pain_data %>%
  group_by(treatment) %>%
  summarise(
    shapiro_p = round(shapiro.test(relief_score)$p.value, 4),
    is_normal = ifelse(shapiro_p > 0.05, "Normal", "Not Normal")
  )

cat("Normality Tests (Shapiro-Wilk):\n")
print(normality_check)

# Visual normality check
par(mfrow = c(1, 3))
for(treat in levels(pain_data$treatment)) {
  subset_data <- pain_data$relief_score[pain_data$treatment == treat]
  qqnorm(subset_data, main = paste("Q-Q:", treat))
  qqline(subset_data, col = "red")
}
par(mfrow = c(1, 1))

# Equal variances test
levene_test <- car::leveneTest(relief_score ~ treatment, data = pain_data)
cat("\nLevene's Test p-value:", round(levene_test$`Pr(>F)`[1], 4))
cat(ifelse(levene_test$`Pr(>F)`[1] > 0.05, " (Equal variances)", " (Unequal variances)"), "\n")

# Step 5: Perform Kruskal-Wallis Test
cat("\n=== Kruskal-Wallis H Test ===\n")

# Main test
kw_result <- kruskal.test(relief_score ~ treatment, data = pain_data)
print(kw_result)

# Manual calculation for understanding
cat("\nManual Calculation Details:\n")
# Overall ranks calculate করি
pain_data$ranks <- rank(pain_data$relief_score)  # ties handled automatically

# Each group এর rank sum
rank_sums <- pain_data %>%
  group_by(treatment) %>%
  summarise(
    n = n(),
    rank_sum = sum(ranks),
    mean_rank = round(mean(ranks), 2)
  )
print(rank_sums)

# Expected rank sum for each group (if no difference)
total_n <- nrow(pain_data)
expected_mean_rank <- (total_n + 1) / 2
cat("Expected mean rank (if no difference):", expected_mean_rank, "\n")

# Step 6: Effect size calculation
# Epsilon-squared for Kruskal-Wallis
H_statistic <- kw_result$statistic
k <- length(levels(pain_data$treatment))  # number of groups
epsilon_squared <- (H_statistic - k + 1) / (total_n - k)

cat("\nEffect Size (Epsilon-squared):", round(epsilon_squared, 3))
if(epsilon_squared < 0.01) {
  cat(" (Small effect)\n")
} else if(epsilon_squared < 0.08) {
  cat(" (Medium effect)\n")
} else {
  cat(" (Large effect)\n")
}

# Step 7: Post-hoc tests (if significant)
if(kw_result$p.value < 0.05) {
  cat("\n=== Post-hoc Tests ===\n")
  cat("Kruskal-Wallis is significant, conducting pairwise comparisons...\n")
  
  # Method 1: Dunn's test with Bonferroni correction
  cat("\n--- Dunn's Test (Bonferroni correction) ---\n")
  dunn_result <- dunn.test(pain_data$relief_score, 
                          pain_data$treatment, 
                          method = "bonferroni",  # correction method
                          alpha = 0.05)
  
  # Method 2: Pairwise Wilcoxon tests with correction
  cat("\n--- Pairwise Wilcoxon Tests ---\n")  
  pairwise_result <- pairwise.wilcox.test(pain_data$relief_score,
                                         pain_data$treatment,
                                         p.adjust.method = "bonferroni")
  print(pairwise_result)
  
  # Method 3: Using FSA package for better formatting
  cat("\n--- Dunn's Test (FSA package) ---\n")
  dunn_fsa <- dunnTest(relief_score ~ treatment, data = pain_data, method = "bonferroni")
  print(dunn_fsa)
  
} else {
  cat("\nKruskal-Wallis is not significant (p >= 0.05)")
  cat("\nNo need for post-hoc tests\n")
}

# Step 8: Compare with ANOVA (educational purpose)
cat("\n=== Comparison with ANOVA ===\n")
anova_result <- aov(relief_score ~ treatment, data = pain_data)
anova_p <- summary(anova_result)[[1]]["treatment", "Pr(>F)"]

cat("ANOVA p-value:", round(anova_p, 6), "\n")
cat("Kruskal-Wallis p-value:", round(kw_result$p.value, 6), "\n")

if(abs(anova_p - kw_result$p.value) < 0.01) {
  cat("Both tests give similar results\n")
} else {
  cat("Tests give different results - check assumptions!\n")
}

# Step 9: Results interpretation
cat("\n=== Results Summary ===\n")
cat("H statistic:", round(kw_result$statistic, 4), "\n")
cat("Degrees of freedom:", kw_result$parameter, "\n") 
cat("p-value:", round(kw_result$p.value, 6), "\n")

if(kw_result$p.value < 0.05) {
  cat("\nConclusion: Significant difference exists among treatments (p < 0.05)\n")
  cat("We reject the null hypothesis.\n")
  cat("At least one treatment shows significantly different pain relief.\n")
  
  # Which group has highest median?
  best_treatment <- desc_stats$treatment[which.max(desc_stats$median)]
  cat("Treatment with highest median relief score:", best_treatment, "\n")
  
} else {
  cat("\nConclusion: No significant difference among treatments (p >= 0.05)\n")
  cat("We fail to reject the null hypothesis.\n") 
  cat("All treatments show similar pain relief effectiveness.\n")
}
```

## 5. Chi-square Test - Categorical Data এর জন্য

### সাধারণ ধারণা:
Chi-square test ব্যবহার করা হয় categorical variables এর জন্য। দুই ধরনের Chi-square test আছে:
1. **Goodness-of-fit:** একটি variable এর distribution expected distribution এর সাথে match করে কিনা
2. **Test of independence:** দুটি categorical variable এর মধ্যে association আছে কিনা

### কখন ব্যবহার করবেন:
- Categorical data (nominal বা ordinal)
- Frequency counts বা proportions নিয়ে কাজ
- দুটি variables এর মধ্যে relationship check করতে

### Real-life উদাহরণ:
**চিকিৎসা:** লিঙ্গ এবং disease occurrence এর মধ্যে সম্পর্ক
**শিক্ষা:** Educational background এবং job satisfaction এর সম্পর্ক  
**ব্যবসা:** Marketing channel এবং customer response এর সম্পর্ক

### Q&A সেশন:

**প্রশ্ন ১:** Expected frequency কিভাবে calculate করি?
**উত্তর:** (Row total × Column total) / Grand total

**প্রশ্ন ২:** Expected frequency কত কম হলে Fisher's exact test ব্যবহার করব?
**উত্তর:** যদি 20% এর বেশি cell এ expected frequency < 5 হয় বা কোনো cell এ < 1 হয়।

**প্রশ্ন ৩:** Cramér's V কী?
**উত্তর:** Chi-square এর effect size measure। 0-1 এর মধ্যে, যেখানে 0 = no association, 1 = perfect association।

```r
# Chi-square Test - Practical R Example
# Gender এবং Product Preference এর মধ্যে association দেখব

# প্রয়োজনীয় libraries
library(ggplot2)
library(dplyr)
library(vcd)        # association measures এর জন্য
library(corrplot)   # visualization এর জন্য

# Step 1: Categorical data তৈরি করি (contingency table)
set.seed(456)

# Simulate survey data: Gender এবং Product Preference
# ধরি একটি survey হয়েছে যেখানে 400 জন participant ছিল
n_male <- 200
n_female <- 200

# পুরুষদের preference (কিছুটা Sports এর দিকে biased)
male_pref <- sample(c("Electronics", "Sports", "Books", "Fashion"), 
                   n_male, replace = TRUE, 
                   prob = c(0.35, 0.40, 0.15, 0.10))  # Sports এর প্রতি বেশি ঝোঁক

# মহিলাদের preference (Fashion এবং Books এর দিকে বেশি)  
female_pref <- sample(c("Electronics", "Sports", "Books", "Fashion"),
                     n_female, replace = TRUE,
                     prob = c(0.20, 0.15, 0.30, 0.35))  # Fashion/Books এর প্রতি বেশি

# Data frame তৈরি করি
survey_data <- data.frame(
  Gender = c(rep("Male", n_male), rep("Female", n_female)),
  Product_Preference = c(male_pref, female_pref)
)

# Step 2: Contingency table তৈরি করি
cat("=== Contingency Table ===\n")
cont_table <- table(survey_data$Gender, survey_data$Product_Preference)
print(cont_table)

# Row percentages দেখি (প্রতি gender এর মধ্যে preference distribution)
cat("\nRow Percentages (Within Gender):\n")
row_percent <- round(prop.table(cont_table, margin = 1) * 100, 1)
print(row_percent)

# Column percentages দেখি (প্রতি product category তে gender distribution)
cat("\nColumn Percentages (Within Product Category):\n") 
col_percent <- round(prop.table(cont_table, margin = 2) * 100, 1)
print(col_percent)

# Step 3: Data visualization
# Stacked bar chart
survey_data %>%
  count(Gender, Product_Preference) %>%           # frequency count
  ggplot(aes(x = Gender, y = n, fill = Product_Preference)) +
  geom_bar(stat = "identity", position = "stack") +    # stacked bars
  labs(title = "Product Preference by Gender (Stacked)",
       x = "Gender", y = "Count", fill = "Product") +
  theme_minimal()

# Grouped bar chart (easier to compare)
survey_data %>%
  count(Gender, Product_Preference) %>%
  ggplot(aes(x = Product_Preference, y = n, fill = Gender)) +
  geom_bar(stat = "identity", position = "dodge") +   # side by side bars
  labs(title = "Product Preference by Gender (Grouped)",
       x = "Product Category", y = "Count", fill = "Gender") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))

# Mosaic plot (proportions visualize করার জন্য best)
mosaicplot(cont_table, 
          main = "Mosaic Plot: Gender vs Product Preference",
          xlab = "Gender", ylab = "Product Preference",
          color = TRUE)

# Step 4: Chi-square test assumptions check
cat("\n=== Chi-square Test Assumptions ===\n")

# Expected frequencies calculate করি
expected_freq <- chisq.test(cont_table)$expected
cat("Expected Frequencies:\n")
print(round(expected_freq, 2))

# Assumption check: Expected frequency >= 5 in 80% cells
cells_less_than_5 <- sum(expected_freq < 5)
total_cells <- length(expected_freq)
percent_less_than_5 <- (cells_less_than_5 / total_cells) * 100

cat("\nCells with expected frequency < 5:", cells_less_than_5, "out of", total_cells)
cat(" (", round(percent_less_than_5, 1), "%)\n")

if(percent_less_than_5 <= 20 && min(expected_freq) >= 1) {
  cat("✓ Assumptions met: Chi-square test is appropriate\n")
} else {
  cat("✗ Assumptions violated: Consider Fisher's exact test\n")
}

# Step 5: Perform Chi-square test
cat("\n=== Chi-square Test of Independence ===\n")

# Main chi-square test
chi_result <- chisq.test(cont_table)
print(chi_result)

# Manual calculation for understanding
cat("\nManual Calculation:\n")
observed <- as.matrix(cont_table)
expected <- chi_result$expected

# Chi-square statistic = Σ((Observed - Expected)² / Expected)
chi_manual <- sum((observed - expected)^2 / expected)
cat("Manual χ² calculation:", round(chi_manual, 4), "\n")
cat("R function result:", round(chi_result$statistic, 4), "\n")

# Degrees of freedom
df <- (nrow(cont_table) - 1) * (ncol(cont_table) - 1)
cat("Degrees of freedom:", df, "\n")

# Step 6: Effect size measures
cat("\n=== Effect Size Measures ===\n")

# Cramér's V (most common for Chi-square)
n_total <- sum(cont_table)
cramers_v <- sqrt(chi_result$statistic / (n_total * (min(dim(cont_table)) - 1)))
cat("Cramér's V:", round(cramers_v, 3))

# Effect size interpretation
if(cramers_v < 0.1) {
  cat(" (Negligible association)\n")
} else if(cramers_v < 0.3) {
  cat(" (Small association)\n") 
} else if(cramers_v < 0.5) {
  cat(" (Medium association)\n")
} else {
  cat(" (Large association)\n")
}

# Phi coefficient (for 2x2 tables only)
if(all(dim(cont_table) == 2)) {
  phi <- sqrt(chi_result$statistic / n_total)
  cat("Phi coefficient:", round(phi, 3), "\n")
}

# Contingency coefficient
cont_coeff <- sqrt(chi_result$statistic / (chi_result$statistic + n_total))
cat("Contingency coefficient:", round(cont_coeff, 3), "\n")

# Step 7: Standardized residuals analysis
cat("\n=== Standardized Residuals ===\n")
std_residuals <- chi_result$stdres
print(round(std_residuals, 2))

cat("\nInterpretation of Standardized Residuals:\n")
cat("Values > |2| indicate significant contribution to Chi-square\n")

# যে cells contribute করেছে বেশি সেগুলো identify করি
significant_cells <- which(abs(std_residuals) > 2, arr.ind = TRUE)
if(nrow(significant_cells) > 0) {
  cat("Cells with significant contribution (|residual| > 2):\n")
  for(i in 1:nrow(significant_cells)) {
    row_idx <- significant_cells[i, 1]
    col_idx <- significant_cells[i, 2]
    row_name <- rownames(cont_table)[row_idx]  
    col_name <- colnames(cont_table)[col_idx]
    residual_val <- std_residuals[row_idx, col_idx]
    
    cat(row_name, "-", col_name, ": residual =", round(residual_val, 2))
    if(residual_val > 2) {
      cat(" (Higher than expected)\n")
    } else {
      cat(" (Lower than expected)\n") 
    }
  }
}

# Step 8: Post-hoc analysis (pairwise comparisons if more than 2x2)
if(min(dim(cont_table)) > 2) {
  cat("\n=== Post-hoc Analysis ===\n")
  cat("For tables larger than 2x2, consider pairwise Chi-square tests\n")
  
  # Example: Compare pairs of product categories
  products <- colnames(cont_table)
  cat("Pairwise comparisons between product categories:\n")
  
  for(i in 1:(ncol(cont_table)-1)) {
    for(j in (i+1):ncol(cont_table)) {
      # Extract 2x2 subtable
      subtable <- cont_table[, c(i, j)]
      pair_test <- chisq.test(subtable)
      
      cat(products[i], "vs", products[j], 
          ": χ² =", round(pair_test$statistic, 3),
          ", p =", round(pair_test$p.value, 4), "\n")
    }
  }
}

# Step 9: Alternative: Fisher's exact test (if assumptions violated)
if(percent_less_than_5 > 20 || min(expected_freq) < 1) {
  cat("\n=== Fisher's Exact Test (Alternative) ===\n")
  cat("Chi-square assumptions violated, using Fisher's exact test...\n")
  
  if(all(dim(cont_table) == 2)) {
    fisher_result <- fisher.test(cont_table)
    print(fisher_result)
  } else {
    cat("Fisher's exact test for tables larger than 2x2 is computationally intensive\n")
    cat("Consider using Monte Carlo simulation:\n")
    fisher_sim <- fisher.test(cont_table, simulate.p.value = TRUE, B = 10000)
    print(fisher_sim)
  }
}

# Step 10: Results interpretation
cat("\n=== Results Summary ===\n")
cat("χ² statistic:", round(chi_result$statistic, 4), "\n")
cat("Degrees of freedom:", chi_result$parameter, "\n")
cat("p-value:", round(chi_result$p.value, 6), "\n")
cat("Cramér's V:", round(cramers_v, 3), "\n")

if(chi_result$p.value < 0.05) {
  cat("\nConclusion: Significant association exists between Gender and Product Preference (p < 0.05)\n")
  cat("We reject the null hypothesis of independence.\n")
  
  # সবচেয়ে strong association কোথায় সেটা identify করি
  max_residual_idx <- which(abs(std_residuals) == max(abs(std_residuals)), arr.ind = TRUE)
  max_row <- rownames(cont_table)[max_residual_idx[1, 1]]
  max_col <- colnames(cont_table)[max_residual_idx[1, 2]]
  
  cat("Strongest association: ", max_row, " and ", max_col, "\n")
  
} else {
  cat("\nConclusion: No significant association between Gender and Product Preference (p >= 0.05)\n")
  cat("We fail to reject the null hypothesis of independence.\n")
  cat("Gender and Product Preference appear to be independent.\n")
}
```

# Part 2: Dependent (Paired) Tests - বিস্তারিত টিউটোরিয়াল

## 6. Paired t-test - Before-After Comparison

### সাধারণ ধারণা:
Paired t-test ব্যবহার করা হয় যখন একই subjects এর উপর দুটি ভিন্ন সময়ে বা condition এ measurement নেওয়া হয়। এটি differences এর mean শূন্য কিনা test করে।

### কখন ব্যবহার করবেন:
- একই subject এর before-after measurement
- Matched pairs (twins, siblings)
- Same subject, different conditions
- Repeated measures (শুধু দুটি time points)

### Real-life উদাহরণ:
**চিকিৎসা:** চিকিৎসার আগে-পরে blood pressure
**শিক্ষা:** training এর আগে-পরে test scores
**ব্যবসা:** marketing campaign এর আগে-পরে sales

### Independent vs Paired t-test:
- **Independent:** দুটি আলাদা গ্রুপ (A গ্রুপ vs B গ্রুপ)
- **Paired:** একই গ্রুপ, দুটি সময় (Before vs After)

### Q&A সেশন:

**প্রশ্ন ১:** Paired t-test কেন বেশি powerful?
**উত্তর:** Individual differences control হয়ে যায়। Each person acts as their own control, তাই variability কম হয়।

**প্রশ্ন ২:** Missing data থাকলে কী করব?
**উত্তর:** Complete pairs নিয়ে কাজ করতে হবে। যদি before আছে কিন্তু after নেই, সেই case exclude করি।

**প্রশ্ন ৩:** Normality assumption কোথায় check করব?
**উত্তর:** Differences এর মধ্যে (after - before), raw data তে নয়।

```r
# Paired t-test - Practical R Example
# Weight loss program এর effectiveness দেখব (before-after study)

# প্রয়োজনীয় libraries
library(ggplot2)
library(dplyr)

# Step 1: Paired data তৈরি করি
set.seed(789)
n_participants <- 30

# Before weight (baseline) - realistic weight distribution  
before_weight <- rnorm(n_participants, mean = 80, sd = 12)  # kg তে

# Weight loss effect simulate করি
# Program কাজ করে, গড়ে ৪ kg weight loss হয়, কিন্তু individual variation আছে
weight_loss <- rnorm(n_participants, mean = 4, sd = 2.5)   # positive = loss

# After weight calculate করি
after_weight <- before_weight - weight_loss

# Data frame তৈরি করি
weight_data <- data.frame(
  participant_id = 1:n_participants,
  before = before_weight,
  after = after_weight,
  difference = after_weight - before_weight  # negative = weight loss
)

# Step 2: Descriptive statistics
cat("=== Descriptive Statistics ===\n")
cat("Before weight - Mean:", round(mean(weight_data$before), 2), 
    ", SD:", round(sd(weight_data$before), 2), "\n")
cat("After weight - Mean:", round(mean(weight_data$after), 2), 
    ", SD:", round(sd(weight_data$after), 2), "\n")
cat("Mean difference (After - Before):", round(mean(weight_data$difference), 2), 
    ", SD:", round(sd(weight_data$difference), 2), "\n")
cat("Mean weight loss:", round(-mean(weight_data$difference), 2), "kg\n")

# Participants who gained vs lost weight
weight_gainers <- sum(weight_data$difference > 0)  
weight_losers <- sum(weight_data$difference < 0)
cat("Participants who lost weight:", weight_losers, "/", n_participants, "\n")
cat("Participants who gained weight:", weight_gainers, "/", n_participants, "\n")

# Step 3: Data visualization
# Before-After comparison using paired data visualization
# Method 1: Paired points with connecting lines
weight_long <- reshape2::melt(weight_data[, c("participant_id", "before", "after")], 
                             id.vars = "participant_id",
                             variable.name = "time", 
                             value.name = "weight")

ggplot(weight_long, aes(x = time, y = weight)) +
  geom_line(aes(group = participant_id), alpha = 0.5, color = "gray") +  # connecting lines
  geom_point(aes(color = time), size = 2, alpha = 0.7) +                # points
  stat_summary(fun = mean, geom = "point", size = 4, color = "red") +    # group means
  stat_summary(fun = mean, geom = "line", aes(group = 1), 
               color = "red", size = 1.2) +                              # mean line
  labs(title = "Weight Before and After Program (Paired Data)",
       x = "Time", y = "Weight (kg)", color = "Time") +
  theme_minimal()

# Method 2: Difference plot (more informative for paired data)
ggplot(weight_data, aes(x = participant_id, y = difference)) +
  geom_hline(yintercept = 0, linetype = "dashed", color = "red") +       # zero line
  geom_point(color = "steelblue", size = 2) +                           # individual differences
  geom_smooth(method = "lm", se = TRUE, color = "darkred") +            # trend line
  labs(title = "Weight Difference (After - Before) by Participant",
       x = "Participant ID", 
       y = "Weight Difference (kg)\n(Negative = Weight Loss)",
       subtitle = paste("Mean difference:", round(mean(weight_data$difference), 2), "kg")) +
  theme_minimal()

# Method 3: Histogram of differences
ggplot(weight_data, aes(x = difference)) +
  geom_histogram(bins = 10, fill = "lightblue", color = "black", alpha = 0.7) +
  geom_vline(xintercept = 0, linetype = "dashed", color = "red", size = 1) +     # zero line
  geom_vline(xintercept = mean(weight_data$difference), 
             linetype = "solid", color = "darkblue", size = 1) +                 # mean line
  labs(title = "Distribution of Weight Differences",
       x = "Weight Difference (kg)", y = "Frequency",
       subtitle = "Red line = No change, Blue line = Mean difference") +
  theme_minimal()

# Step 4: Assumptions check for paired t-test
cat("\n=== Paired t-test Assumptions Check ===\n")

# Assumption 1: Normality of DIFFERENCES (not raw data)
shapiro_diff <- shapiro.test(weight_data$difference)
cat("Normality test of differences (Shapiro-Wilk):\n")
cat("p-value:", round(shapiro_diff$p.value, 4))
cat(ifelse(shapiro_diff$p.value > 0.05, " (Normal)", " (Not Normal)"), "\n")

# Visual normality check - Q-Q plot of differences
qqnorm(weight_data$difference, main = "Q-Q Plot: Weight Differences")
qqline(weight_data$difference, col = "red")

# Assumption 2: Independence of differences (check by study design)
cat("Independence: Each participant measured independently ✓\n")

# Assumption 3: No extreme outliers in differences
# Using IQR method
Q1_diff <- quantile(weight_data$difference, 0.25)
Q3_diff <- quantile(weight_data$difference, 0.75)
IQR_diff <- Q3_diff - Q1_diff
outlier_threshold_low <- Q1_diff - 1.5 * IQR_diff
outlier_threshold_high <- Q3_diff + 1.5 * IQR_diff

outliers <- weight_data$difference < outlier_threshold_low | 
           weight_data$difference > outlier_threshold_high
n_outliers <- sum(outliers)

cat("Outliers in differences:", n_outliers, "/", n_participants, "\n")
if(n_outliers > 0) {
  cat("Outlier values:", round(weight_data$difference[outliers], 2), "\n")
}

# Step 5: Perform Paired t-test
cat("\n=== Paired t-test Results ===\n")

# Method 1: Using paired argument
paired_result <- t.test(weight_data$after, weight_data$before, paired = TRUE)
print(paired_result)

# Method 2: Direct test on differences (equivalent)
diff_result <- t.test(weight_data$difference, mu = 0)  # test if mean difference = 0
cat("\nAlternative method (testing differences directly):\n")
cat("Same t-statistic:", round(diff_result$statistic, 4), "\n")
cat("Same p-value:", round(diff_result$p.value, 6), "\n")

# Step 6: Manual calculation for understanding
cat("\n=== Manual Calculation ===\n")
mean_diff <- mean(weight_data$difference)
sd_diff <- sd(weight_data$difference)
se_diff <- sd_diff / sqrt(n_participants)  # standard error of differences
t_manual <- mean_diff / se_diff            # t-statistic

cat("Mean difference:", round(mean_diff, 4), "\n")
cat("SD of differences:", round(sd_diff, 4), "\n") 
cat("SE of differences:", round(se_diff, 4), "\n")
cat("t-statistic (manual):", round(t_manual, 4), "\n")
cat("df:", n_participants - 1, "\n")

# Step 7: Effect size calculation
# Cohen's d for paired data = mean_diff / sd_diff
cohens_d_paired <- mean_diff / sd_diff
cat("\nEffect Size (Cohen's d):", round(cohens_d_paired, 3))

if(abs(cohens_d_paired) < 0.2) {
  cat(" (Small effect)\n")
} else if(abs(cohens_d_paired) < 0.5) { 
  cat(" (Small to medium effect)\n")
} else if(abs(cohens_d_paired) < 0.8) {
  cat(" (Medium to large effect)\n") 
} else {
  cat(" (Large effect)\n")
}

# Step 8: Confidence interval interpretation
cat("\n=== Confidence Interval ===\n")
ci_lower <- paired_result$conf.int[1]
ci_upper <- paired_result$conf.int[2]

cat("95% CI for mean difference: [", round(ci_lower, 2), ",", round(ci_upper, 2), "]\n")
cat("95% CI for mean weight loss: [", round(-ci_upper, 2), ",", round(-ci_lower, 2), "] kg\n")

if(ci_lower > 0) {
  cat("CI suggests significant weight GAIN\n")
} else if(ci_upper < 0) {
  cat("CI suggests significant weight LOSS\n") 
} else {
  cat("CI includes zero - no significant change\n")
}

# Step 9: Compare with independent t-test (educational purpose)
cat("\n=== Comparison with Independent t-test ===\n")
# Treating as independent samples (WRONG approach for paired data)
independent_wrong <- t.test(weight_data$after, weight_data$before, paired = FALSE)

cat("Paired t-test p-value:", round(paired_result$p.value, 6), "\n")
cat("Independent t-test p-value:", round(independent_wrong$p.value, 6), "\n")
cat("Paired test is", round(independent_wrong$p.value / paired_result$p.value, 1), 
    "times more powerful\n")

# Step 10: Results interpretation
cat("\n=== Results Summary ===\n") 
cat("Sample size:", n_participants, "participants\n")
cat("Mean weight change:", round(mean_diff, 2), "kg")
cat(ifelse(mean_diff < 0, " (Loss)", " (Gain)"), "\n")
cat("t-statistic:", round(paired_result$statistic, 4), "\n")
cat("df:", paired_result$parameter, "\n")
cat("p-value:", round(paired_result$p.value, 6), "\n")

if(paired_result$p.value < 0.05) {
  cat("\nConclusion: Significant weight change occurred (p < 0.05)\n")
  cat("We reject the null hypothesis (no change).\n")
  
  if(mean_diff < 0) {
    cat("The weight loss program was EFFECTIVE.\n")
    cat("Average weight loss:", round(-mean_diff, 2), "kg per participant.\n")
  } else {
    cat("Participants significantly GAINED weight.\n") 
    cat("Average weight gain:", round(mean_diff, 2), "kg per participant.\n")
  }
} else {
  cat("\nConclusion: No significant weight change (p >= 0.05)\n")
  cat("We fail to reject the null hypothesis.\n")
  cat("The weight loss program showed no significant effect.\n")
}
```

## 7. Wilcoxon Signed-rank Test - Paired t-test এর Non-parametric বিকল্প

### সাধারণ ধারণা:
Wilcoxon signed-rank test হলো paired t-test এর non-parametric version। এটি differences এর ranks ব্যবহার করে কাজ করে এবং normality assumption এর প্রয়োজন নেই।

### কখন ব্যবহার করবেন:
- Paired data আছে কিন্তু differences নরমাল নয়
- Ordinal scale data
- Extreme outliers আছে
- Small sample size এবং normality uncertain

### Real-life উদাহরণ:
**চিকিৎসা:** Pain rating (1-10 scale) before-after treatment
**শিক্ষা:** Student satisfaction survey before-after course  
**Psychology:** Mood rating scale before-after intervention

### কিভাবে কাজ করে:
1. Differences calculate করে
2. Zero differences remove করে
3. Absolute differences এর rank করে  
4. Original sign দিয়ে signed ranks তৈরি করে
5. Positive এবং negative ranks এর sum তুলনা করে

### Q&A সেশন:

**প্রশ্ন ১:** Tied values থাকলে কী হয়?
**উত্তর:** R automatically handles করে average ranks assign করে।

**প্রশ্ন ২:** Zero differences এর সাথে কী করি?
**উত্তর:** Typically exclude করি, কিন্তু কিছু variant আছে যেগুলো include করে।

**প্রশ্ন ৩:** Effect size কিভাবে calculate করি?
**উত্তর:** r = Z/√N, যেখানে Z হলো test statistic এর z-score approximation।
